```
import numpy as np
import matplotlib.pyplot as plt
```

```
X = np.arange(0.1, 16, 0.001)
K = 0.8
Y = np.sin(K*X) / X / K
dY = (np.sin(K*X) - K*X*np.cos(K*X)) / (K *X *X)
X_axis = np.zeros(X.shape)
Y2 = np.zeros(X.shape)


plt.plot(X,Y)
plt.plot(X, dY)
plt.plot(X, X_axis)
plt.show()
```

```
start_merge = False
val = -1
eps = 1e-9
minimum_val = 0
minimum_index = 0
Y2 = np.zeros(X.shape)

for i in range(X.shape[0]):
  if not start_merge :
    Y2[i] = Y[i]
    if minimum_val > Y[i]:
      minimum_val = Y[i]
      minimum_index = i

    if X[i] > 12 and np.abs(dY[i]) < 1e-3:
      val = Y[i]
      start_merge = True
  else:
    Y2[i] = val

for i in range(minimum_index):
  Y2[i] = minimum_val

plt.plot(X,Y2)
plt.show()
```

```
from mpl_toolkits import mplot3d

X = np.outer(np.linspace(-10,10,100), np.ones(100))
Y = X.T

Z1 = (1 -(X*X + Y*Y) / 16) / 3

Z2 = np.sin( (X + Y + 10) ) / 50


for i in range(Z1.shape[0]):
  for j in range(Z1.shape[1]):
    if Z1[i, j] < 0 :
      Z1[i, j] = 0

Z = Z1 + Z2
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(X,Y,Z)
ax.set_zlim([0, 1])
plt.show()



```
